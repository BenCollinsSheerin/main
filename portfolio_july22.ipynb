```python
import numpy as np
import matplotlib.pyplot as plt

asset_value = np.array([2000, 5000, 500, 500, 500, 500, 1000, 1000])
asset_labels = ["NA", "Cash (GBP)", "Natural Gas (USD)", "Grains (USD)", "Physical Gold (USD)", "Physical Silver (USD)", "Short EUR Long USD", "Short GBP Long USD"]

plt.title("Winter of Discontent Portfolio", loc = 'center', size=16)
plt.pie(asset_value, startangle=90, wedgeprops={"edgecolor":"k",'linewidth': 2, 'linestyle': 'solid', 'antialiased': True})
plt.legend(labels = asset_labels, loc='right', bbox_to_anchor=(1.4, 0.5))
plt.show()
```
